# Creating token embeddings

* Step 1: Define the inputs and dimensions First, you define your sample token IDs along with a simplified vocabulary size and the desired dimension for the output embeddings

```python
import torch

# A sample sequence of four token IDs
input_ids = torch.tensor([2, 3, 5, 1])

# A simplified vocabulary of 6 words, and a target embedding size of 3 dimensions
vocab_size = 6
output_dim = 3
```

* Step 2: Initialize the embedding layer Next, you initialize the embedding layer using torch.nn.Embedding. Setting a manual random seed ensures the initially random weights are identical every time you run the code.

```python

# setting random weights (using a fixed seed for reproducibility)
torch.manual_seed(123)

# create the embedding layer
embedding_layer = torch.nn.Embedding(vocab_size, output_dim)
"""
1. It creates the underlying weight matrix PyTorch allocates memory and constructs a new weight matrix (a tensor). 
It uses the two arguments you provided to define the exact shape of this matrix:
Rows (vocab_size): It creates one row for every possible token in your vocabulary
Columns (output_dim): It creates a column for every dimension you want your output vector to have
If your vocab_size is 6 and your output_dim is 3, PyTorch builds a 6x3 matrix in memory

2. It populates the matrix with random values PyTorch doesn't leave this matrix empty; it automatically fills it with small, random values. 
If you ran a command like torch.manual_seed(123) immediately before initializing the layer, PyTorch uses that seed to ensure the "random" values generated for this matrix are exactly the same every time you run your script

3. It prepares the matrix for future training By default, PyTorch sets this embedding weight matrix as a trainable parameter (specifically, it sets requires_grad=True).
This tells PyTorch's autograd engine that these random numbers are just the starting point, and they are allowed to be optimized and updated later during the backpropagation process when you actually begin training the model
"""
# Optional: You can print the layer's underlying weight matrix (6 rows x 3 columns)
print(embedding_layer.weight)
```

* Step 3: Convert the sequence of token IDs Finally, you pass your sequence of token IDs directly into the embedding layer to perform the lookup operation

```python
# Convert the discrete token IDs into continuous embedding vectors
print(embedding_layer(input_ids))
# Crucially, this step only translates the tokens into a continuous format that the deep neural network can process mathematically
# Neither backpropagation nor weight adjustments occur during this specific execution call.
```

We have got a 4x3 matrix, meaning your four input tokens have successfully been converted into four 3-dimensional embedding vectors
