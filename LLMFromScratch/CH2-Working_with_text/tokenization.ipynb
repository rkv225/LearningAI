## Tokenizing a Text Sample
This notebook demonstrates reading a text file and converting it into tokens for natural language processing.

## Read the Text File
Load the text data into a variable for processing.

```python
with open("the-verdict.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
print("Total number of character:", len(raw_text))
print(raw_text[:99])
```

## Tokenize the Text
Convert the text into a list of tokens using NLTK.

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(raw_text)
print("Total number of tokens:", len(tokens))
print(tokens[:20])
```

## Build the Vocabulary
Create a sorted list of unique tokens from the text.

```python
all_words = sorted(set(tokens))
print("Total number of unique tokens:", len(all_words))
```

## Create the Token-to-Index Mapping
Assign a unique integer to each token.

```python
# Add special tokens for unknown and end of text
vocab = {token: integer for integer, token in enumerate(all_words)}
vocab['<|unk|>'] = len(vocab)
vocab['<|endoftext|>'] = len(vocab)

# Example: append end of text token to tokens
# tokens.append('<|endoftext|>')

for i, item in enumerate(vocab.items()):
    print(item)
    if i >= 20:
        break
```

## Tokenizer Class
Define a simple tokenizer for encoding and decoding text.

```python
class SimpleTokenizer:
    def __init__(self, vocab, eos_token="<|endoftext|>"):
        self.str_to_int = vocab
        self.int_to_str = {integer: token for token, integer in vocab.items()}
        self.eos_token = eos_token

    def encode(self, text, add_eos=True):
        preprocessed_text = word_tokenize(text)
        if add_eos:
            preprocessed_text.append(self.eos_token)
        ids = [self.str_to_int.get(token, self.str_to_int['<|unk|>']) for token in preprocessed_text]
        return ids

    def decode(self, ids):
        tokens = [self.int_to_str.get(id, '<|unk|>') for id in ids]
        text = " ".join(tokens)
        return text
```

## Encode and Decode Example
Demonstrate encoding and decoding a sample text using the tokenizer.

```python
tokenizer = SimpleTokenizer(vocab)
sample_text = "Poor Jack Gisburn! unknownword"
encoded = tokenizer.encode(sample_text)
print("Encoded:", encoded[:20])
decoded = tokenizer.decode(encoded)
print("Decoded:", decoded)
```
